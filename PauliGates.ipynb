```python
from ibm_quantum_widgets import CircuitComposer
from qiskit import qiskit, QuantumRegister, ClassicalRegister, QuantumCircuit
from numpy import pi

qc = QuantumCircuit(3, 3)
#not gate practice
qc.x(0)
qc.measure(0, 0)

job = qiskit.execute(qc, qiskit.BasicAer.get_backend('qasm_simulator'))
print(job.result().get_counts())
```

```python
#CNOT gate
qc.cx(0, 1)
qc.measure(1,1)
#And/CCNOT/Toffoli gate
qc.ccx(0, 1, 2)
qc.measure(2, 2)

job = qiskit.execute(qc, qiskit.BasicAer.get_backend('qasm_simulator'))
print(job.result().get_counts())
```

```python
#NAND
qc.x(2)
qc.measure(2,2)

job = qiskit.execute(qc, qiskit.BasicAer.get_backend('qasm_simulator'))
print(job.result().get_counts())
```

```python
#x,y,y->h measure/phase visualization
qc = QuantumCircuit(3, 3)
qc.x(0)
qc.y(1)
qc.y(2)
qc.h(2)
qc.measure(range(3), range(3))
job = qiskit.execute(qc, qiskit.BasicAer.get_backend('qasm_simulator'))
print(job.result().get_counts())
```

```python
#XOR
qc.reset(range(3))
qc.x(0)
qc.x(1)
qc.cx(0,2)
qc.cx(1,2)
qc.measure(range(3), range(3))
job = qiskit.execute(qc, qiskit.BasicAer.get_backend('qasm_simulator'))
print(job.result().get_counts())
```

```python
#OR
qc.reset(range(3))
qc.x(0)
qc.x(1)
qc.cx(0,2)
qc.cx(1,2)
qc.ccx(0,1,2)
qc.measure(range(3), range(3))
job = qiskit.execute(qc, qiskit.BasicAer.get_backend('qasm_simulator'))
print(job.result().get_counts())
```
